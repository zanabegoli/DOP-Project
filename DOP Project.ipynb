## Data loading

```python
import pandas as pd
```

```python
political_data = pd.read_csv('datasets/political_rights.csv')
political_data
```

## creating index

```python
countries = political_data.location.unique()
index = []
for country in countries:
    for year in range(2003, 2017):
        index.append((country, year))
index = pd.MultiIndex.from_tuples(index, names= ['Country', 'Year'])
```

## preparing data shape

```python
all_data = pd.DataFrame(columns=['Political Rights', 'Civil Liberties'], index=index)
all_data
```

## adding political data to all data

```python
for index, row in political_data.iterrows():
    if row['variable'].strip() == 'Political Rights':
        all_data.loc[(row['location'], row['Date']), 'Political Rights'] = row['Value']
    if row['variable'].strip() == 'Civil Liberties':
        all_data.loc[(row['location'], row['Date']), 'Civil Liberties'] = row['Value']
all_data
```
